```python
import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Health Care Anaysis/Data Set.csv")
```

```python
df.head()
```

<h1>3 Financial Analysis</h1>
<h3>3.1 Billing Amount Analysis</h3>
<h4>3.1.1 Distribution Analysis</h4>

```python
# Calculate summary statistics
billing_stats = df['Billing Amount'].describe()
mean_billing = df['Billing Amount'].mean()
median_billing = df['Billing Amount'].median()
std_billing = df['Billing Amount'].std()

print("Summary Statistics:")
print(billing_stats)
print(f"Mean Billing Amount: {mean_billing:.2f}")
print(f"Median Billing Amount: {median_billing:.2f}")
print(f"Standard Deviation: {std_billing:.2f}")
```

```python
# Create the histogram
plt.figure(figsize=(10, 10))
hist_plot = sns.histplot(df['Billing Amount'], bins=30, kde=True, color='green')

# Add the height of each bar on top of the bar with rotated text
for rect in hist_plot.patches:
    height = rect.get_height()
    if height > 0:  # Only add text for bars with height greater than 0
        plt.text(
            rect.get_x() + rect.get_width() / 2,  # x position
            height,  # y position
            f'{int(height)}',  # label text
            ha='center', va='bottom', fontsize=10, color='black',
            rotation=90  # Rotate the text by 90 degrees
        )

plt.title('Distribution of Billing Amounts')
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Billing Amount', data=df, color='skyblue')
plt.title('Box Plot of Billing Amounts')
plt.xlabel('Billing Amount')
plt.show() 
```

<h4>3.1.2 Correlation Analysis</h4>

```python
# Select numerical columns and encode categorical variables
df_encoded = df.copy()
df_encoded['Medical Condition'] = df_encoded['Medical Condition'].astype('category').cat.codes
df_encoded['Admission Type'] = df_encoded['Admission Type'].astype('category').cat.codes
df_encoded['Insurance Provider'] = df_encoded['Insurance Provider'].astype('category').cat.codes

# Correlation matrix
corr_matrix = df_encoded[['Billing Amount', 'Medical Condition', 'Admission Type', 'Insurance Provider']].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Billing Amount and Other Variables')
plt.show()
```

<h3>3.2 Insurance Provider Analysis</h3>
<h4>3.2.1 Billing Amount Comparison</h4>

```python
# Grouping by Insurance Provider and calculating average billing amounts
insurance_billing = df.groupby('Insurance Provider')['Billing Amount'].mean().reset_index()

# Sorting by average billing amount in descending order
highest_billing_provider = insurance_billing.sort_values(by='Billing Amount', ascending=False).head(1)

# print("Insurance Provider with the Highest Average Billing Amount:")
# print(highest_billing_provider)

# Create the box plot
plt.figure(figsize=(12, 8))
ax = sns.boxplot(x='Insurance Provider', y='Billing Amount', data=df, palette='coolwarm')

# Highlight the insurance provider with the highest average billing amount
highest_provider = highest_billing_provider['Insurance Provider'].values[0]
for patch in ax.artists:
    provider = patch.get_label()
    if provider == highest_provider:
        patch.set_edgecolor('red')
        patch.set_linewidth(2)

plt.title('Billing Amounts by Insurance Provider')
plt.xlabel('Insurance Provider')
plt.ylabel('Billing Amount')
plt.xticks(rotation=45)
plt.show()
```

<h4>3.2.2 Condition or Hospital-Specific Analysis</h4>

```python
# Group by Insurance Provider and Medical Condition to see the count
insurance_condition = df.groupby(['Insurance Provider', 'Medical Condition']).size().unstack().fillna(0)

# Heatmap for Insurance Provider vs Medical Condition
plt.figure(figsize=(12, 8))
sns.heatmap(insurance_condition, cmap='coolwarm', annot=True, fmt=".0f")
plt.title('Heatmap of Insurance Providers vs Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Insurance Provider')
plt.show()
```

```python
# Group by Hospital and calculate the average billing amount
hospital_billing = df.groupby('Hospital')['Billing Amount'].mean().reset_index()
# Sort by average billing amount and select the top 10 hospitals
top_10_hospitals = hospital_billing.sort_values(by='Billing Amount', ascending=False).head(10)['Hospital']
# print(top_10_hospitals)
# Filter the dataframe to include only the top 10 hospitals
df_top_10_hospitals = df[df['Hospital'].isin(top_10_hospitals)]
# Group by Insurance Provider and Hospital and count occurrences
insurance_hospital = df_top_10_hospitals.groupby(['Insurance Provider', 'Hospital']).size().unstack().fillna(0)

# Create the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(insurance_hospital, cmap='coolwarm', annot=True, fmt=".0f")
plt.title('Heatmap of Insurance Providers vs Top 10 Hospitals')
plt.xlabel('Hospital')
plt.ylabel('Insurance Provider')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()

```
