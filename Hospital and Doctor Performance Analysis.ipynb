<h1>2... Hospital and Doctor Performance</h1>

<h3>2a... Doctors Performance Analysis</h3>

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Convert necessary columns to appropriate data types
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'], errors='coerce')
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'], errors='coerce')

# Example: Analyzing the performance of doctors based on test results
doctor_performance = df.groupby(['Doctor', 'Test Results']).size().unstack(fill_value=0)

# Plotting the performance for a sample of doctors
sample_doctors = doctor_performance.sample(n=30, random_state=1)  # Choose a random sample of 10 doctors
sample_doctors.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Test Results by Doctor')
plt.xlabel('Doctor')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.show()

# # Analyze discharge status by doctor
# discharge_status_by_doctor = df.groupby(['Doctor', 'Discharge Date']).size().unstack(fill_value=0)
# print(discharge_status_by_doctor)
```

<h3>2a1... Doctor Performance Based on Patient Outcomes</h3>
<h4>Discharge Status and Test Results</h4>

```python
# Grouping data by Doctor and calculating the percentage of successful outcomes
doctor_performance = df.groupby('Doctor').agg(
    total_patients=('Test Results', 'count'),
    normal_tests=('Test Results', lambda x: (x == 'Normal').sum()),
    inconclusive_tests=('Test Results', lambda x: (x == 'Inconclusive').sum()),
    abnormal_tests=('Test Results', lambda x: (x == 'Abnormal').sum())
)

# Calculating percentages
doctor_performance['normal_percentage'] = (doctor_performance['normal_tests'] / doctor_performance['total_patients']) * 100
doctor_performance['inconclusive_percentage'] = (doctor_performance['inconclusive_tests'] / doctor_performance['total_patients']) * 100
doctor_performance['abnormal_percentage'] = (doctor_performance['abnormal_tests'] / doctor_performance['total_patients']) * 100

# Displaying the results
doctor_performance.reset_index()
```

<h4>Condition-Specific Doctor Analysis</h4>

```python
# Grouping data by Doctor and Medical Condition
condition_specific_analysis = df.groupby(['Doctor', 'Medical Condition']).agg(
    total_patients=('Test Results', 'count'),
    normal_tests=('Test Results', lambda x: (x == 'Normal').sum()),
    abnormal_tests=('Test Results', lambda x: (x == 'Abnormal').sum())
)

# Calculating percentages
condition_specific_analysis['normal_percentage'] = (condition_specific_analysis['normal_tests'] / condition_specific_analysis['total_patients']) * 100
condition_specific_analysis['abnormal_percentage'] = (condition_specific_analysis['abnormal_tests'] / condition_specific_analysis['total_patients']) * 100

# Displaying the results
condition_specific_analysis.reset_index()
```

<h3>2b... Hospital Performance</h3>
<h4>Patient Outcomes by Hospital</h4>

```python
# Grouping data by Hospital
hospital_performance = df.groupby('Hospital').agg(
    total_patients=('Test Results', 'count'),
    normal_tests=('Test Results', lambda x: (x == 'Normal').sum()),
    inconclusive_tests=('Test Results', lambda x: (x == 'Inconclusive').sum()),
    abnormal_tests=('Test Results', lambda x: (x == 'Abnormal').sum())
)

# Calculating percentages
hospital_performance['normal_percentage'] = (hospital_performance['normal_tests'] / hospital_performance['total_patients']) * 100
hospital_performance['inconclusive_percentage'] = (hospital_performance['inconclusive_tests'] / hospital_performance['total_patients']) * 100
hospital_performance['abnormal_percentage'] = (hospital_performance['abnormal_tests'] / hospital_performance['total_patients']) * 100

# Displaying the results
hospital_performance.reset_index()
```

<h4>Billing Amount Comparison by Hospital</h4>

```python
# Grouping by Hospital and Admission Type to analyze cost differences
billing_analysis = df.groupby(['Hospital', 'Admission Type']).agg(
    average_billing=('Billing Amount', 'mean'),
    total_patients=('Billing Amount', 'count')
).reset_index()

# Displaying the results
billing_analysis
```

```python
import matplotlib.pyplot as plt

# Grouping by Hospital and calculating the average billing amount
billing_analysis = df.groupby('Hospital').agg(
    average_billing=('Billing Amount', 'mean'),
    total_patients=('Billing Amount', 'count')
).reset_index()

# Sorting by average billing amount and selecting the top 10 hospitals
top_10_hospitals = billing_analysis.sort_values(by='average_billing', ascending=False).head(10)

# Plotting the data
plt.figure(figsize=(26, 8))
bars = plt.barh(top_10_hospitals['Hospital'], top_10_hospitals['average_billing'], color='skyblue')

# Adding the average billing values on top of each bar
for bar in bars:
    plt.text(
        bar.get_width() + 500,  # Position of the text (slightly to the right of the bar)
        bar.get_y() + bar.get_height() / 2,  # Centered vertically with respect to the bar
        f'{bar.get_width():,.2f}',  # Text to display (formatted with commas and two decimals)
        va='center',  # Vertical alignment
        ha='left'  # Horizontal alignment
    )

plt.xlabel('Average Billing Amount')
plt.ylabel('Hospital')
plt.title('Top 10 Hospitals by Average Billing Amount')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value on top
plt.show()
```
